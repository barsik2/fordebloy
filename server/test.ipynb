```python
import pandas as pd
import numpy as np
from faker import Faker

# Инициализация Faker
fake = Faker('ru_RU')

# Количество людей
num_people = 100

# Создание списка типов
types = [1, 2, 3, 4, 5, 6]

# Функция для генерации данных для одного человека
def generate_person_data():
    return {
        'id': fake.unique.random_int(min=1, max=num_people),
        'type': np.random.choice(types),
        'fio': fake.name(),
        'nickname': fake.user_name(),
        'email': fake.email(),
        'phone': fake.phone_number(),
        'job': fake.job()
    }

# Создание DataFrame с людьми и их данными
people_data = [generate_person_data() for _ in range(num_people)]
people = pd.DataFrame(people_data)

# Генерация случайных связей между людьми
connections = []
people_ids = people['id'].tolist()
for person_id in people_ids:
    num_connections = np.random.randint(1, 5)  # от 1 до 4 связей
    available_ids = [pid for pid in people_ids if pid != person_id]  # доступные ID для связи
    connected_to = np.random.choice(available_ids, num_connections, replace=False)
    connection_types = np.random.choice([1, 2, 3], num_connections)
    
    for conn_id, conn_type in zip(connected_to, connection_types):
        connections.append({'from': person_id, 'to': conn_id, 'connection_type': conn_type})

connections_df = pd.DataFrame(connections)

# Вывод DataFrame с людьми
print("People DataFrame:")
print(people.head())

# Вывод DataFrame со связями
print("\nConnections DataFrame:")
print(connections_df.head())
```

```python
import pandas as pd
import numpy as np
from faker import Faker

# Инициализация Faker
fake = Faker('ru_RU')

# Количество людей
num_people = 100

# Определение типов и их вероятности
types = [1, 2, 3, 4, 5, 6]
type_labels = {
    1: 'Employee',
    2: 'Manager',
    3: 'Director',
    4: 'Partner',
    5: 'Client',
    6: 'Support Staff'
}
type_probabilities = [0.4, 0.2, 0.1, 0.05, 0.2, 0.05]  # Вероятности для каждого типа

# Определение типов связей и их вероятности
connection_types = [1, 2, 3]
connection_labels = {
    1: 'Business Collaboration',
    2: 'Friendship',
    3: 'Supervision'
}
connection_probabilities = [0.5, 0.3, 0.2]  # Вероятности для каждого типа связи

# Определение весов для каждого типа связи в виде векторов
connection_weights = {
    1: [np.random.rand(3).tolist() for _ in range(3)],  # Веса для делового сотрудничества
    2: [np.random.rand(3).tolist() for _ in range(3)],  # Веса для дружеских отношений
    3: [np.random.rand(3).tolist() for _ in range(3)]   # Веса для руководства/подчинения
}
weight_probabilities = [0.5, 0.3, 0.2]  # Вероятности для весов

# Функция для генерации данных для одного человека
def generate_person_data():
    person_type = np.random.choice(types, p=type_probabilities)
    return {
        'id': fake.unique.random_int(min=1, max=num_people * 10),  # чтобы избежать дублирования id
        'type': person_type,
        'type_label': type_labels[person_type],
        'fio': fake.name(),
        'nickname': fake.user_name(),
        'email': fake.email(),
        'phone': fake.phone_number(),
        'job': fake.job()
    }

# Создание DataFrame с людьми и их данными
people_data = [generate_person_data() for _ in range(num_people)]
people = pd.DataFrame(people_data)

# Генерация случайных связей между людьми
connections = []
people_ids = people['id'].tolist()
for person_id in people_ids:
    num_connections = np.random.randint(1, 5)  # от 1 до 4 связей
    available_ids = [pid for pid in people_ids if pid != person_id]  # доступные ID для связи
    connected_to = np.random.choice(available_ids, num_connections, replace=False)
    connection_types_selected = np.random.choice(connection_types, num_connections, p=connection_probabilities)
    
    for conn_id, conn_type in zip(connected_to, connection_types_selected):
        weight = np.random.choice(connection_weights[conn_type], p=weight_probabilities)
        connections.append({
            'from': person_id,
            'to': conn_id,
            'connection_type': conn_type,
            'connection_label': connection_labels[conn_type],
            'weight': weight
        })

connections_df = pd.DataFrame(connections)

# Вывод DataFrame с людьми
print("People DataFrame:")
print(people.head())

# Вывод DataFrame со связями
print("\nConnections DataFrame:")
print(connections_df.head())
```
